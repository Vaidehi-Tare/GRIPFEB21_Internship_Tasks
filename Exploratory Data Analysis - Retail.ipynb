> ## Bivariate Analysis

### 1. With respect to 'Sales'

```python
plt.figure(figsize = (6,6))
plt.title('Regionwise Sales of the Superstore')
plt.pie(store['Sales'].groupby(by = store.Region).sum(), 
        labels = store['Sales'].groupby(by = store.Region).sum().index,
        autopct = '%.2f%%')
my_circle = plt.Circle((0,0), 0.8, color = 'white')
p = plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

```python
plt.figure(figsize = (10,7))
plt.title('Sales vs Category of Products')
sns.barplot(x = 'Category', y = 'Sales', data = store, palette = 'cool')
```

```python
plt.figure(figsize = (20,8))
plt.title('Sales vs Sub-Category of Products')
sns.barplot(x = 'Sub-Category', y = 'Sales', data = store, palette = 'plasma')
plt.show()
```

```python
plt.figure(figsize = (6,6))
plt.pie(store['Sales'].groupby(by = store['Segment']).sum(), 
        labels =store['Sales'].groupby(by = store['Segment']).sum().index,
        autopct = '%.3f%%', explode = (0.1,0,0))
plt.title('Segment wise Sales of the products')
plt.show()
```

### 2. With respect to 'Profit'

```python
plt.figure(figsize = (6,6))
plt.title('Regionwise Profit of the Superstore')
plt.pie(store['Profit'].groupby(by = store.Region).sum(), 
        labels = store['Profit'].groupby(by = store.Region).sum().index,
        autopct = '%.2f%%')
my_circle = plt.Circle((0,0), 0.8, color = 'white')
p = plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

```python
plt.figure(figsize = (10,7))
plt.title('Profit vs Category of Products')
sns.barplot(x = 'Category' , y = 'Profit', data = store, palette = 'cool')
plt.show()
```

```python
plt.figure(figsize = (20,8))
plt.title('Profit vs Sub-Category of Products')
sns.barplot(x = 'Sub-Category', y = 'Profit', data = store, palette = 'plasma')
plt.show()
```

```python
plt.figure(figsize = (6,6))
plt.pie(store['Profit'].groupby(by = store['Segment']).sum(), 
        labels =store['Profit'].groupby(by = store['Segment']).sum().index,
        autopct = '%.3f%%', explode = (0.1,0,0))
plt.title('Segment wise Profit of the products')
plt.show()
```

> ## Insights derived from Bivariate Analysis

* The **Western Region** has the **most Sales** contributing about **32% of the total Sales** 
* The **Southern Region** has the **lowest Sales** contributing about **17% of the total Sales**
* Products under the **Category of 'Technology' generate the maximum Sales**
* The **Top 5 selling products** are from the following Sub-Category - **Copiers, Machines, Tables, Chairs and Bookcases**
* **Consumer Segments contributes the most to total sales** as compared to the other 2 segments
* The **Western Region** also contributes the **most in total Profit (with about 38% out of the total Profit)** while the **Central Region** contributes **the least to the Profit(with only 13% out of the total Profit)**
* Products under the **Category of 'Technology' generates the maximum Profit**
* Out of the **Top 5 selling Sub-Categories, only Copiers, Machines and Chairs have managed to earn Profit**
* **Consumer Segment** contributes the most to Profit as well along with Sales

> ## Multivariate Analysis

```python
store[['Profit', 'Quantity']].groupby(by = store['Region']).sum().sort_values(by = 'Profit', ascending = False)
```

```python
plt.figure(figsize = (15,8))
plt.title('Sales of each Category in each Ship Mode')
sns.barplot(x = 'Category', y = 'Sales', data = store, hue = store['Ship Mode'], palette = 'pastel')
plt.show()
```

```python
pd.DataFrame(store['Quantity'].groupby(by = [store['Category'], store['Region']]).count()).unstack().plot(kind = 'bar', 
                                                                                                          stacked = True,
                                                                                                         figsize = (10,8))
plt.title('Quantity of products in each Category for each region')
plt.show()
```

```python
plt.figure(figsize = (10,8))
plt.title('Correlation between Sales, Profit, Quantity and Discount')
sns.heatmap(store.corr(), annot = True, cmap = 'YlGnBu')
plt.show()
```

```python
sns.pairplot(data = store.drop('Postal Code', axis = 1))
plt.show()
```

> ## Insights derived from Multivariate Analysis

* The Profit and Quantity of products sold Region wise can be seen clearly
* Products under the **category of 'Technology'** has the **highest sales** and the **preferred Ship Mode is 'Same Day'.** **'Second Class'** ship mode is the **least preferred** for products under Technology
* **Highest quantity** of products sold belong to the **Office Supplies** and they are sold the **most in the Western Region** followed by the Eastern Region
* Products sold **least in terms of Quantity** are Technology related products
* **Sales is highly correlated with Profit and Quantity** while **Discount** is positively correlated to **Quantity**
